```python
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------
# Global style settings
# ------------------------
plt.rcParams.update({
    "font.size": 18,          # default font size
    "axes.labelsize": 20,     # axis label font size
    "axes.titlesize": 22,     # title font size
    "xtick.labelsize": 18,    # x-axis tick font size
    "ytick.labelsize": 18,    # y-axis tick font size
    "legend.fontsize": 18,    # legend font size
    "lines.linewidth": 3,     # default line width
    "lines.markersize": 8     # marker size
})

# ------------------------
# Data
# ------------------------
steps = [
    10, 20, 30, 40, 43,
    53, 63, 73, 83, 86, 93, 103, 113, 123, 128, 133, 143, 153, 163, 170, 173, 183, 193, 203, 212,
    222, 226, 232, 239, 242, 252, 262, 265, 272, 278, 282, 291, 292, 302, 304, 312, 317, 322, 330, 342,
]

scores = [
    0.605,0.629,0.627,0.655,0.6424,
    0.655,0.643,0.646,0.659,0.648,0.668,0.641,0.651,0.66,0.6506,0.648,0.67,0.667,0.661,0.66,0.661,
    0.652,0.662,0.679,0.6742,0.687,0.6806,0.686,0.6822,0.688,0.691,0.703,0.6854,0.696,0.6878,0.683,
    0.6984,0.713,0.696,0.7006,0.72,0.72,0.722,0.7236,0.722
]

df = pd.DataFrame({"step": steps, "score": scores})

# ------------------------
# Plotting
# ------------------------
plt.figure(figsize=(12,6))

# Gray background line to ensure continuity
plt.plot(df["step"], df["score"], color="gray", alpha=0.3, zorder=1)

# Stage splits
warmup_df = df[df["step"] <= 43]
distractor_df = df[(df["step"] > 43) & (df["step"] <= 212)]
filter_df = df[df["step"] > 212]

# Draw stage-specific curves
plt.plot(warmup_df["step"], warmup_df["score"], marker="o", color="#1f77b4", label="Warmup", zorder=2)
plt.plot(distractor_df["step"], distractor_df["score"], marker="o", color="#ff7f0e", label="Distractor", zorder=2)
plt.plot(filter_df["step"], filter_df["score"], marker="o", color="#2ca02c", label="Hard-Mined", zorder=2)

# Vertical lines for stage transitions
for xpos, name in [(43, "Warmup  Distractor"), (212, "Distractor  Hard-Mined")]:
    plt.axvline(x=xpos, linestyle="--", alpha=0.6, color="gray")
    plt.text(xpos + 1, plt.gca().get_ylim()[1]*0.93, name,
             va="top", color="black", fontsize=18)

# Axis labels
plt.xlabel("Total training steps")
plt.ylabel("LongBench1 average score")
plt.legend()
plt.grid(True, alpha=0.3)

# Save to file
plt.savefig("qwen-7b_lb1_curve.pdf", bbox_inches="tight")
plt.show()
```
