```python
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------
# 
# ------------------------
steps = [
    # Series A
    10, 20, 30, 40, 43,
    # Series B
    53, 63, 73, 83, 86, 93, 103, 113, 123, 128, 133, 143, 153, 163, 170, 173, 183, 193, 203, 212,
    # Series C
    222, 226, 232, 239, 242, 252, 262, 265, 272, 278, 282, 291, 292, 302, 304, 312, 317, 322, 330
]

scores = [
    0.605,0.629,0.627,0.655,0.6424,
    0.655,0.643,0.646,0.659,0.648,0.668,0.641,0.651,0.66,0.6506,0.648,0.67,0.667,0.661,0.66,0.661,
    0.652,0.662,0.679,0.6742,0.687,0.6806,0.686,0.6822,0.688,0.691,0.703,0.6854,0.696,0.6878,0.683,
    0.6984,0.713,0.696,0.7006,0.72,0.72,0.722,0.7236
]

df = pd.DataFrame({"step": steps, "score": scores})

# ------------------------
# 
# ------------------------
plt.figure(figsize=(12,6))

# 
warmup_df = df[df["step"] <= 43]
distractor_df = df[(df["step"] > 43) & (df["step"] <= 212)]
filter_df = df[df["step"] > 212]

# 
plt.plot(warmup_df["step"], warmup_df["score"], marker="o", color="#1f77b4", label="Warmup")
plt.plot(distractor_df["step"], distractor_df["score"], marker="o", color="#ff7f0e", label="Distractor")
plt.plot(filter_df["step"], filter_df["score"], marker="o", color="#2ca02c", label="Filter")

# 
for xpos, name in [(43, "Warmup  Distractor"), (212, "Distractor  Filtered")]:
    plt.axvline(x=xpos, linestyle="--", alpha=0.5, color="gray")
    plt.text(xpos + 1, plt.gca().get_ylim()[1]*0.98, name,
             fontsize=9, va="top", color="black")

plt.xlabel("Total training steps")
plt.ylabel("LongBench1 average score")
plt.title("Training Progress vs LongBench1 Score")
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
